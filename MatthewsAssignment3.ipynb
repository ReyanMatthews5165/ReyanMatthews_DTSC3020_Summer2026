## Question 3 — Sets, Set Operations, and Boolean Values  *(13 points)*

Use sets to manage unique items and compare groups, and use Boolean values to check conditions.

### What to do
1. A sign-in list with repeated names is provided: `attendees = ["Anna", "John", "Sara", "Anna", "John", "Lina"]`. Convert it to a **set** to keep only the unique names, store it in `unique_attendees`, and print both the original list and the set.
2. Two groups are provided in the code cell:
   - `python_students = {"Anna", "John", "Sara", "Reza"}`
   - `data_students = {"Sara", "Reza", "Lina", "Omid"}`
   
   Print each of the following, labeled:
   - all unique students in **either** group (`union`)
   - students in **both** groups (`intersection`)
   - students in `python_students` but **not** in `data_students` (`difference`)
3. **Boolean values:** A student can enter the lab if they are **at least 18 years old** AND have a **student ID**. Using the variables `age = 20` and `has_student_id = True`, create a Boolean called `can_enter_lab` that is `True` only when **both** conditions are met, and print it.
4. **Membership and subset:** Create a set `project_team = {"Sara", "Reza"}`. Print the result of `project_team.issubset(python_students)`, and print the result of `"Anna" in python_students`.
5. Add a **comment** answering: what is the difference between `union` and `intersection`? (One short sentence.)

### Hints
- Make a set from a list with `set(my_list)` — duplicates are removed automatically.
- `a.union(b)`, `a.intersection(b)`, and `a.difference(b)` compare two sets.
- `and` requires **both** sides to be `True`.
- `issubset()` returns `True` when every item of the first set is in the second.

````python
# Re'yan Matthews - Question 3

# 1. Convert a list to a set
attendees = ["Anna", "John", "Sara", "Anna", "John", "Lina"]

unique_attendees = set(attendees)

print("Original list:", attendees)
print("Unique attendees:", unique_attendees)

# 2. Set operations
python_students = {"Anna", "John", "Sara", "Reza"}
data_students = {"Sara", "Reza", "Lina", "Omid"}

print("Union:", python_students.union(data_students))
print("Intersection:", python_students.intersection(data_students))
print("Difference:", python_students.difference(data_students))

# 3. Boolean values
age = 20
has_student_id = True

can_enter_lab = age >= 18 and has_student_id

print("Can enter lab:", can_enter_lab)

# 4. Membership and subset
project_team = {"Sara", "Reza"}

print("Project team is a subset of python students:",
      project_team.issubset(python_students))
print("Anna in python_students:", "Anna" in python_students)

# 5. Comment
# Union combines all items from both sets, while intersection includes only items found in both sets.
````

## Question 4 — Dictionaries and List Comprehension  *(12 points)*

Store key–value data in a dictionary and use list comprehension to pull information out of it.

### What to do
A nested dictionary of students is provided in the code cell. Each student has a `grade` and a list of `courses`:

```python
students = {
    "Anna": {"grade": 92, "courses": ["Python", "Statistics"]},
    "John": {"grade": 76, "courses": ["Python", "Math"]},
    "Sara": {"grade": 88, "courses": ["AI", "Python"]},
    "Lina": {"grade": 65, "courses": ["Math", "Statistics"]},
    "Reza": {"grade": 95, "courses": ["AI", "Python"]}
}
```

1. **Access:** Print Anna's grade, and print Sara's first course (the item at index 0 of her course list).
2. **Add and update:** Add a new student `"Mike"` with a grade of `81` and courses `["Python", "AI"]`. Then update Lina's grade to `70`. Print the full dictionary afterward.
3. **List comprehension — passing students:** Assume a passing grade is **70 or higher**. Use a list comprehension to build a list called `passed_students` containing the **names** of all students who passed, then print it.
4. **List comprehension — by course:** Use a list comprehension to build a list called `python_students` containing the **names** of all students who have `"Python"` in their course list, then print it.
5. **Slicing:** Print only the **first three** names from `passed_students` using slicing.
6. Add a **comment** answering: what does `students.items()` give you when you loop over it? (One short sentence.)

### Hints
- Access a nested value with two keys: `students["Anna"]["grade"]`.
- Add or update with `students["Mike"] = {...}` and `students["Lina"]["grade"] = 70`.
- List comprehension format: `[name for name, info in students.items() if info["grade"] >= 70]`.
- Check membership in a list with `"Python" in info["courses"]`.
- Slicing: `passed_students[:3]` gives the first three items.

````python
# Re'yan Matthews - Question 4

students = {
    "Anna": {"grade": 92, "courses": ["Python", "Statistics"]},
    "John": {"grade": 76, "courses": ["Python", "Math"]},
    "Sara": {"grade": 88, "courses": ["AI", "Python"]},
    "Lina": {"grade": 65, "courses": ["Math", "Statistics"]},
    "Reza": {"grade": 95, "courses": ["AI", "Python"]}
}

# 1. Access values
print("Anna's grade:", students["Anna"]["grade"])
print("Sara's first course:", students["Sara"]["courses"][0])

# 2. Add and update
students["Mike"] = {"grade": 81, "courses": ["Python", "AI"]}
students["Lina"]["grade"] = 70

print("Updated dictionary:", students)

# 3. Passing students
passed_students = [
    name for name, info in students.items()
    if info["grade"] >= 70
]

print("Passed students:", passed_students)

# 4. Students taking Python
python_students = [
    name for name, info in students.items()
    if "Python" in info["courses"]
]

print("Python students:", python_students)

# 5. First three passing students
print("First three passed students:", passed_students[:3])

# 6. Comment
# students.items() returns each student's name and information as a key-value pair.

````
